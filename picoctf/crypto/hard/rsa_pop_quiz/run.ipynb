```python
def gcd(a, b): # b > a
    if a>b:
        temp = a
        a = b
        b = temp
    q_0 = 0
    a_0 = a
    b_0 = b
    ### extended gcd ###
    s_0 = 1
    s_1 = 0
    t_0 = 0
    t_1 = 1
    ### extended gcd ###
    while True:
        q_0 = 0
        while b_0>=a_0: #  and a_0!=0
            b_0 -= a_0
            q_0+=1
        ### extended gcd ###
        temp_s = s_1
        s_1 = s_0-q_0*s_1
        s_0 = temp_s
        temp_t = t_1
        t_1 = t_0-q_0*t_1
        t_0 = temp_t
        # print(t_0, t_1, s_0, s_1)
        ### extended gcd ###
        if (q_0==0 or b_0==0): break
        rem = b_0
        b_0 = a_0
        a_0 = rem
        if (a_0==1):
            temp_s = s_1
            s_1 = s_0-q_0*s_1
            s_0 = temp_s
            temp_t = t_1
            t_1 = t_0-q_0*t_1
            t_0 = temp_t
            break
    return a_0, t_0, s_0
def lcm(a,b):
    return abs(a*b)//gcd(a,b)[0]
ascii32bit = "abcdefghijklmnopqrstuvwxyz.,!?' "
def ascii32bit_encode(s: str):
    acc = 0
    i = 0
    for c in s:
        acc+=(ascii32bit.index(c) if c in ascii32bit else 31)<<i
        i+=5
    return acc
# edge case where last char has leading zeros, need to adjust bit shift offset accordingly
def ascii32bit_decode(s: int):
    acc = s
    ans = ""
    n = round(log2(s)) - 5
    n = n+(5-n%5) if n%5!=0 else n
    for i in range(n, -1, -5):
        temp = acc >> i
        temp2 = temp << i
        ans=ascii32bit[temp]+ans
        acc -= temp2
    return ans
def fast_mod_exp(b, exp, m):
    res = 1
    while exp > 1:
        if exp & 1:
            res = (res * b) % m
        b = b ** 2 % m
        exp >>= 1
    return (b * res) % m
# ported from https://stackoverflow.com/questions/2921406/calculate-primes-p-and-q-from-private-exponent-d-public-exponent-e-and-the
# given d e and n, find p and q
def pq_from_den(d, e, n):
    k = d*e-1
    if (k%2==0):
        r = k
        t = 0
        while (r%2==0):
            r = r >> 1
            t+=1
        success = False
        y = -1
        for i in range(1,101):
            g = -1
            g = random.randrange(0,n)
            y = fast_mod_exp(g,r,n)
            if y==1 or y==n-1:
                continue
            for j in range(1, t):
                x = fast_mod_exp(y,2,n)
                if x==1:
                    success = True
                    break
                if x==n-1:
                    continue
                y = x
            x = fast_mod_exp(y, 2, n)
            if (x==1):
                success=True
                break
        if (success):
            p = gcd(y-1, n)[0]
            q = n//p
            return p, q
        else:
            assert False, "not found"
    else:
        assert False, "not found"
```
