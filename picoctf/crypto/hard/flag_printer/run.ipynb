```python
points = []
for line in open('encoded.txt', 'r').read().strip().split('\n'):
    p, q = line.split(' ')
    points.append((int(p), int(q)))
GF = galois.GF(N)
```

```python
matrix = np.empty((n,n), dtype=np.uint64)
solution = np.empty(n, dtype=np.uint64)
row = np.empty(n, dtype=np.uint64)
j = 0
for point in points[0:n]:
    p, q = point
    solution[j] = q
    acc = 1
    for i in range(n): # len(points)
        matrix[j][i] = (acc % N)
        acc*=p
    # matrix.append(GF(row))
    j+=1
```

```python
out = np.linalg.solve(GF(matrix), GF(solution))
```
