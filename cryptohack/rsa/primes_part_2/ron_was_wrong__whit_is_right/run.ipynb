```python
from Crypto.PublicKey import RSA
from Crypto.Util.number import GCD, isPrime, bytes_to_long, inverse, long_to_bytes
N = []
C = []
e = 65537
for i in range(1,51):
    n = RSA.import_key(open(f"keys_and_messages/{str(i)}.pem", 'r').read()).n
    C.append(bytes_to_long(bytes.fromhex(open(f"keys_and_messages/{str(i)}.ciphertext", 'r').read())))
    N.append(n)
```

```python
for i in range(len(N)):
    for j in range(i):
        if (i==j): continue
        a = GCD(N[i], N[j])
        if (a!=1): print(i, j)
```

```python
from Crypto.Cipher import PKCS1_OAEP
p1 = GCD(N[20], N[33]) # loop over N
p2 = N[20]//p1
p3 = N[33]//p1
d = inverse(e, (p1-1)*(p2-1))
PKCS1_OAEP.new(RSA.construct((N[20], e, d, p1, p2))).decrypt(long_to_bytes(C[20]))
```

```python
for i in range(len(N)):
    if (GCD(N[i], p2*p3)!=1): print(i)
```
