```python
def bitFlip(pos, data):
    bit_pos = pos%8
    byte_pos = (pos-bit_pos)//8
    ret = data
    ret = ret[:byte_pos]+bytes([ret[byte_pos]^(1 << (7-bit_pos))])+ret[byte_pos+1:]
    return ret
```
